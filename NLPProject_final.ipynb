```python
'''
Model Selection: Sentiment Analysis model for binary classification of telugu text into positive or negative sentiment.
source: https://huggingface.co/ravikumar1478/Sentiment_Analysis_on_Telugu_Movie_Reviews/tree/main
'''
```

```python
'''
Tokenization/ Encoding (preprocessing, seq to indices and padding) is model dependent hence the pre-built tokenizer needs to called for encoding the datasets into their encodings.
max_length of encodings is set to 259 here (max_sentence_length)
but the max_length of encodings is 429 here as special labels such as SEP, CLS are added to the sentences
'''
from transformers import AutoTokenizer, AutoModelForSequenceClassification

tokenizer = AutoTokenizer.from_pretrained("ravikumar1478/Sentiment_Analysis_on_Telugu_Movie_Reviews", max_length=max_sentence_length)

X_train = train_data['text'].tolist()
y_train = train_data['label'].tolist()
X_test = test_data['text'].tolist()
y_test = test_data['label'].tolist()
```

```python
train_encodings = tokenizer(X_train, truncation=True, padding=True)
test_encodings = tokenizer(X_test, truncation=True, padding=True)
```

```python
import torch

class CustomDataset(torch.utils.data.Dataset):
    def __init__(self, encodings, labels):
        self.encodings = encodings
        self.labels = labels

    def __getitem__(self, idx):
        item = {key: torch.tensor(val[idx]) for key, val in self.encodings.items()}
        item['labels'] = torch.tensor(self.labels[idx])
        return item

    def __len__(self):
        return len(self.labels)

# Assuming train_encodings, test_encodings are dictionaries containing encoded data
# and y_train, y_test are the corresponding labels
train_dataset = CustomDataset(train_encodings, y_train)
test_dataset = CustomDataset(test_encodings, y_test)
```

#Hyperparameter Tuning -an extension
We shall extend the code for fine tuning our model further by also doing hyperparameter tuning,I have run the code for around 5 iteration and stopped it because it takes around 15*15->225 parameter combinations which takes more than an hour.And found the best hyperparameter after these iterations.

```python
from transformers import TrainingArguments, Trainer, AutoModelForSequenceClassification
from sklearn.model_selection import ParameterGrid

# Define the hyperparameter grid
param_grid = {
    'per_device_train_batch_size': [4, 8, 16],
    'per_device_eval_batch_size': [8, 16, 32],
    'warmup_steps': [100, 500, 1000],
    'weight_decay': [0.01, 0.1],
    'logging_steps': [5, 10, 20]
}

best_loss = float('inf')
best_hyperparameters = None

for params in ParameterGrid(param_grid):
    training_args = TrainingArguments(
        output_dir='./results',
        num_train_epochs=2,
        per_device_train_batch_size=params['per_device_train_batch_size'],
        per_device_eval_batch_size=params['per_device_eval_batch_size'],
        warmup_steps=params['warmup_steps'],
        weight_decay=params['weight_decay'],
        logging_dir='./logs',
        logging_steps=params['logging_steps'],
        save_strategy="steps",
        eval_strategy="steps",
        load_best_model_at_end=True,
        metric_for_best_model='accuracy'
    )

    model = AutoModelForSequenceClassification.from_pretrained("ravikumar1478/Sentiment_Analysis_on_Telugu_Movie_Reviews")

    trainer = Trainer(
        model=model,
        args=training_args,
        train_dataset=train_dataset,
        eval_dataset=test_dataset
    )
    trainer.train()
    eval_result = trainer.evaluate()

    if eval_result['eval_loss'] < best_loss:
        best_loss = eval_result['eval_loss']
        best_hyperparameters = params


```
